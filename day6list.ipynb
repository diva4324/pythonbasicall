```python
#list compression
```

```python
numbers=[2,3,4,5,6]
squares =[x**2 for x in numbers if x%2 ==0]
print(squares)
```

```python
numbers=[2,3,4,5,6]
squares =[x**2 for x in numbers if x%2 !=0]
print(squares)
```

```python
y=[]
for x in numbers:
    if x%2 !=0:
        z= x**2
        y.append(z)
print(y)
```

```python
odd= [x for x in range(10) if x%2 !=0]
print(odd)
```
