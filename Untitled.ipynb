```python
class Employee:
    def __init__(self, name, salary):
        self.__name = name
        self.__salary = salary

    # Getter
    def get_name(self):
        return self.__name

    def get_salary(self):
        return self.__salary

    # Setter
    def set_salary(self, new_salary):
        if new_salary > 0:
            self.__salary = new_salary
        else:
            print("Invalid salary")

```

```python

emp = Employee("John", 50000)

print("Name:", emp.get_name())
print("Old Salary:", emp.get_salary())

```

```python
emp.set_salary(60000)
print("New Salary:", emp.get_salary())
```

```python
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks   

    def calculate_average(self):
        return sum(self.marks) / len(self.marks)

    def calculate_grade(self):
        avg = self.calculate_average()
        if avg >= 90:
            return "A"
        elif avg >= 75:
            return "B"
        elif avg >= 50:
            return "C"
        else:
            return "F"


```

```python
student1 = Student("Pragya", 101, [85, 90, 78, 88, 92])

print("Name    :", student1.name)
print("Roll No :", student1.roll_number)
print("Average :", round(student1.calculate_average(), 2))
print("Grade   :", student1.calculate_grade())
```

```python
steps = []

for i in range(7):
    steps.append(int(input("Enter steps: ")))

total_steps = 0

print("\n---Daily Steps ---")

for i in range(7):
    total_steps += steps[i]
    print(f"Day {i + 1}: {steps[i]} steps")

average = total_steps / 7

print("\n--- Weekly Summary ---")
print(f"Weekly Total : {total_steps} steps")
print(f"Average      : {average:.2f} steps")

```

```python
secret = 1234
attempts = 0

while True:
    pin = int(input("Enter PIN: "))
    attempts += 1

    if pin == secret:
        print("Access Granted")
        break
    elif attempts == 3:
        print("System Locked")
        break
    else:
        print("Wrong PIN, try again.")
```

```python

returned_books = [
    "1984", "Hobbit", "1984", "Dune", "Hobbit",
    "Foundation", "Dune", "Brave New World",
    "1984", "Fahrenheit 451", "Hobbit"
]


unique_books = set(returned_books)


library = {'1984': 5, 'Hobbit': 2, 'Dune': 4}


for book in unique_books:
    if book in library:
        library[book] += 1
    else:
        library[book] = 1


print(library)
```
