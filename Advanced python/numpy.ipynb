#### Slicing and Reshaping Arrays

* Slicing helps extract parts of arrays.
* Reshaping changes the shape without changing data.

```python
import numpy as np

arr = np.arange(12)
print("Original:", arr)

#Slicing
print("First 5:", arr[:5])
print("Last 3:", arr[-3:])

print("3rd to 7th", arr[3:8])
```

```python
#reshape to 3*4
reshaped=arr.reshape(3,4)
print("reshape 3*4:\n", reshaped)
```

##### Vectorized Operations & Broadcasting

* `Vectorized operations` allow fast element-wise operations.
* `Broadcasting` enables operations between arrays of different shapes.

# Two dimensions are compatible when:
They are equal, OR
One of them is 1.

```python
x=np.array([1,2,3])
y=np.array([4,5,6])
print("addition",x+y)
print("multiplication",x*y)
print("scalar multiplication",x*10)
```

```python
# Broadcasting example
mat = np.ones((2, 3), dtype=int)
print(mat)

print("-------")
vec = np.array([1, 2, 3])
print("Broadcasted Addition:\n", mat + vec)
```

```python
vect = np.array([1, 2, 3])
vectr = np.array([[1, 2, 3],[4, 5, 6]])
print("Broadcasted Addition:\n", vect + vectr)
scalar=5
print("Broadcasted scalar multiplication:\n", scalar * vectr)
```

##### Boolean Indexing & Filtering

- Use boolean masks to filter arrays based on conditions.

- In NumPy, you can pass a boolean array inside square brackets to filter the original array. This is called boolean indexing.

```python
scores = np.array([80, 85, 90 ,96,88,78])
high_scores= scores[scores>70]
print("scores>70:",high_scores)
even_scores =scores[scores%2==0]
print("scores>70:",even_scores)
odd_scores =scores[scores%2!=0]
print("scores>70:",odd_scores)
```

## 3. Working with Missing Data in NumPy

Unlike **Pandas**, standard **NumPy integer arrays** do not support missing values (`NaN`).  
To work with missing data, the array usually needs to be of type **float** or **object**.

### ðŸ”¹ Key Concepts

- **`np.nan`**  
  Represents a missing or undefined value (*Not a Number*).

- **`np.isnan()`**  
  Checks which elements are missing.

- **`np.nansum()`**, **`np.nanmean()`**  
  Perform arithmetic operations while safely ignoring `NaN` values.

```python
# Create an array with missing data (must be float)
data = np.array([1.0, 2.0, np.nan, 4.0, 5.0, np.nan])

print(f"Array with NaNs: {data}")

bool_miss =np.isnan(data)
print(f"is nan?: {bool_miss}")
print(f"count of nan: {np.sum(bool_miss)}")
```

```python
print(f"sum of fails: {np.sum(data)}")
```

```python
print(f"safe sum: {np.nansum(data)}")
```

```python
print(f"safe mean: {np.nanmean(data)}")
```

```python
# 5. Removing NaNs
clean_data = data[~np.isnan(data)]
print(f"Data with NaNs removed: {clean_data}")

# 6. Filling NaNs with a value (e.g., 0)
filled_data = np.copy(data)
filled_data[np.isnan(filled_data)] = 0
print(f"Data with NaNs filled with 0: {filled_data}")
```
