OOPs:
Analogy: Building a house
    A class is like a blueprint — it defines the structure and features (walls, doors, rooms).
    An object is a real house built from that blueprint — each with its own color, size, and number of rooms.

OOP helps by:
    Organizing code around real-world concepts
    Making it easier to reuse, extend, and maintain
    Allowing creation of multiple unique instances from a single design

1. Class
    A blueprint for creating objects.
    Like a house plan that defines what every house should have.
    class House:


2. Object (Instance)
    A concrete version of a class with real values.
    A specific house built from the blueprint.
    my_house = House()


3. Attribute
    Data stored in an object.
    Features of a house like color, size, address.
    my_house.color = "blue"


4. Method
    A function inside a class that defines behavior.
    Actions like open_door(), turn_on_lights().
    my_house.open_door()

```python
### Defining Methods

# Methods are functions defined inside a class that define the object's behavior. They also take `self` as the first parameter,
# allowing them to access and modify the object's attributes.


class Dog:
    def __init__(self, name, breed, age):
        self.name = name
        self.breed = breed
        self.age = age
        print(f"A new Dog object named {self.name} was created!")


    def bark(self):
        print(f"{self.name} says: Woof! Woof!")

    def describe(self):
        print(f"This is {self.name}, a {self.age}-year-old {self.breed}.")

    def have_birthday(self):
        self.age += 1
        print(f"Happy Birthday, {self.name}! You are now {self.age} years old.")
```

```python
dog3 = Dog("get", "Golden dog", 4)
dog3.describe()
```

```python
dog3.have_birthday()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"Person object created: {self.name}")

    def introduce(self):
        print(f"Hi, my name is {self.name} and I am {self.age} years old.")

    def walk(self):
        print(f"{self.name} is walking.")
```

```python
class Student(Person):
    pass

s1 = Student("Bob",20)

print(f"studnet name :{s1.name}")
s1.introduce()
s1.walk()
    
```
