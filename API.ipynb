```python
import requests

api_url = "https://jsonplaceholder.typicode.com/todos/2";

# Make the GET request
response = requests.get(api_url)
print(response)
# The response object contains all the information
print(f"Status Code: {response.status_code}") # 200 means OK

# The actual data is in the response body
# .text gives the raw text, .json() parses it directly into a Python dict
if response.status_code == 200:
    data = response.json()
    print("Data from API:")
    print(data)
    print(f"Title: {data['title']}")
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python

import requests
base_currency = input("enter the base currency")

target_currency =input("enter the target currency")
base_currency=base_currency.upper()
target_currency=target_currency.upper()

api_url = f"https://open.er-api.com/v6/latest/{base_currency}"
response = requests.get(api_url)

if response.status_code == 200:
    data = response.json()
    rates = data["rates"]
    
else:
    print("Request failed:", response.status_code)

def convert_currency(base_currency, target_currency, rates):
    value_of_base = rates[base_currency]
    value_of_target = rates[target_currency]
    

    final_value = value_of_base* value_of_target
    return f"Present Currency price of {target_currency} with {base_currency} :{final_value}"
    
final_value=convert_currency(base_currency, target_currency, rates)
print(final_value)
```
