### What are Dictionaries?
Unordered collections of key-value pairs

Keys must be unique and immutable (strings, numbers, tuples)

Values can be any Python object

```python
# Method 1: Using the curly braces {}
empty_dict = {}

student = {
    "name": "Alice",
    "age": 21,
   
    "courses": ["Maths", "Physics"]
}

print(student)
```

```python
student["address"] ="Kathmandu Nepal"
print(student)
```

```python
student.get("age")
#another way
print(student["age"])
print(student.get("courses"[0]))
```

```python
grades=dict([("math",90),("physics",85)]) #another way of indexing
```

```python
employee = dict(
    name="Bob",
    department="HR",
    employee_id=9876,
    years_of_service=5
)
print(employee)
print(grades)
print(type(employee))
```
