### 1. What are Sets?
 Sets are unordered collections of unique, immutable elements.
*   They do not store duplicate elements.
*   Elements must be immutable (like keys in dictionaries).
*   Sets themselves are mutable (you can add/remove elements).
*   There is no concept of order or index.
*   Very efficient for membership testing (`in`) and removing duplicates.

```python
emptyset =set()
print(emptyset)
print(type(emptyset))
```

```python
numbers = {1, 2, 3, 4, 4, 5, 2}
print(numbers)
```

```python
dir(set)
```

```python
thisset = {"apple", "banana", "cherry", True, 1, 2}

print(thisset)
```

```python
thisset = {"apple", "banana", "cherry", False, 0, 2}

print(thisset)
```

```python
colors_list = ["red", "green", "blue", "red", "yellow", "blue"]
colors_set = set(colors_list)
print(colors_set) #list lai set garayera print gareko 
```

```python
chars_set = set("hello world")
print(chars_set)  #space pani aaucha set ma
```

```python
set_a = {1,3,4,5,6,7,8,9}
set_b = {3,4,5,6,1,2,8,22,33,44}
union_set_op = set_a | set_b
union_set_meth = set_a.union(set_b)

print(f"unino(|) :{union_set_op}")
print(f"union(method) :{union_set_meth}")
```

```python

difference_set_meth = set_a.difference(set_b)
```

```python
print(f"differnece :{difference_set_meth}")
```
