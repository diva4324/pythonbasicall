## Abstraction

#### Abstraction means:
* Hiding complicated stuff (like wires inside the controller).
* Showing only whatâ€™s important (the buttons you need to press).

```python
# ABC (Abstract Base Class)
from abc import ABC, abstractmethod

# Create an abstract class
class Animal(ABC):
   
    @abstractmethod
    def make_sound(self):
        pass
   
    @abstractmethod
    def move(self):
        pass
   
    def sleep(self):
        print("Zzz... sleeping")
        return

# Create concrete classes that inherit from Animal
class Dog(Animal):
   
    def make_sound(self):
        print("Woof! Woof!")
   
    def move(self):
        print("Running on four legs")

    def number_of_legs(self):
        print("Have four legs")

class Bird(Animal):
   
    def make_sound(self):
        print("Chirp! Chirp!")
   
    def move(self):
        print("Flying in the sky")

    def number_of_legs(self):
        print("Have 2 legs")
```
