```python
def addandsub(a,b):
    result = a+b
    result1= a-b

    return result, result1
```

```python
addandsub(5,9)
```

```python
x=5
y=9

add_result, sub_result=addandsub(x,y)
print(add_result)
print(sub_result)
```

```python
# *args and *kwargs
#use args to pass multiple arguments

def show_args(*args):
    lists=""
    for item in args:

        print(item)
        lists += item

    return lists
    

value = show_args("1","2","3","4")
print(value)

```

```python
# *args and *kwargs
#use args to pass multiple arguments

def show_args(*args):
    global lists
    lists=0
   
    for item in args:

        print(item)
        lists += int(item)

    return lists
    

value = show_args("1","2","3","4")
print(value)

print(f"{lists}")
```

```python
sum = 0
def addition(*args):
    for item in args:
        global sum
        sum = sum + item
    return sum

addition(1,2,4,22,44,55,55)
```
